```python
"""
PART I - SPECIFIC ASSIGNMENTS
03
Produce a boxplot that shows the distribution of total points, total assists, 
and total rebounds (each of these three is a separate box plot, but they can be on the same scale and in the same graphic).
"""
uri_3 = "https://raw.githubusercontent.com/Alissoninacio1/Data-Analyzis-/main/basketball_master.csv"
master = pd.read_csv(uri_3)
nba = pd.merge(players, master, how="left", left_on="playerID", right_on="bioID")

sns.boxplot(data=nba[["points", "rebounds", "assists"]]) #boxplot of multiple columns using seaborn
```

```python
 ax = sns.boxplot(data=nba[["points", "rebounds", "assists"]], color='y') #boxplot using new arguments
```

```python
"""
PART I - SPECIFIC ASSIGNMENTS
04
Produce a plot that shows how the number of points scored has changed over time by showing the median of points scored per year, over time.
The x-axis is the year and the y-axis is the median number of points among all players for that year.
"""


"""
To plot rebounds per year, we first need to group the statistics by year. 
When we do so, we need to specify how we want to aggregate the data of that year. 
In this case, we'll use the median of the reboundsPerGame
"""
nba_year_set = nba[["points", "year"]].groupby("year").median()
print(nba_year_set)

```

```python
#In order to plot this data, I need to change the index to be the year now
nba_year_set = nba_year_set.reset_index()
sns.regplot(data=nba_year_set, x="year", y="points")
```

```python
#let's remove any years where the median was 0
nba_year_set = nba_year_set[nba_year_set["points"] > 0]
sns.regplot(data=nba_year_set, x="year", y="points").set_title("Median Points per Year among all players")
```

```python
# Another way to consider this is to find the top 10 rebounders of the year and look at their median.
 
 
# Get the top 10 rebounders per year
nba_topPoints_perYear = nba[["points", "year"]].groupby("year")["points"].nlargest(10)

# Get the median of these 10
nba_topPoints_median_perYear = nba_topPoints_perYear.groupby("year").median()

# Put year back in as a column
nba_topPoints_median_perYear = nba_topPoints_median_perYear.reset_index()

# Again no zeros...
nba_topPoints_median_perYear_noZeros = nba_topPoints_median_perYear[nba_topPoints_median_perYear["points"] > 0]

# Now plot
sns.regplot(data=nba_topPoints_median_perYear_noZeros, x="year", y="points").set_title("Median of Top 10 points Each Year")
```
